```python
import random

import numpy as np

from blob import blob_sphere
from ktree.ntree import NTreeDynamic

N_DIMENSION = 3

random.seed(0)

all_data = np.concatenate([
    blob_sphere(100, (0, 0, 0), 100),
    blob_sphere(50, (50, 50, 50), 25),
    blob_sphere(50, (-25, -50, -25), 25),
    blob_sphere(50, (-75, -75, -75), 5),
])

tree = NTreeDynamic(1)
```

```python
for a in all_data:
    tree.insert(a)
```

```python
sorted_data = tree.sort()
sorted_data = sorted(sorted_data, key=lambda x: len(x))[::-1]
```

```python
import random

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)


def plot_clusters(fig, clusters):
    ax1 = fig.add_subplot(2, 3, 1, projection='3d')
    ax2 = fig.add_subplot(2, 3, 2, projection='3d')
    ax3 = fig.add_subplot(2, 3, 3, projection='3d')
    ax4 = fig.add_subplot(2, 3, 4, projection='3d')
    ax5 = fig.add_subplot(2, 3, 5, projection='3d')
    ax6 = fig.add_subplot(2, 3, 6, projection='3d')

    for data in clusters:

        c_data = np.array(data)
        # colors = np.mean(c_data, axis=0)

        colors = random.random(), random.random(), random.random()

        x = data[:, 0]
        y = data[:, 1]
        z = data[:, 2]

        ax1.scatter(x, y, z, c=colors, alpha=0.6, s=20)
        ax1.set_title('Isometric View (30°, -60°)', fontsize=10)
        ax1.view_init(elev=30, azim=-60)

        ax2.scatter(x, y, z, c=colors, alpha=0.6, s=20)
        ax2.set_title('Front View (0°, -90°)', fontsize=10)
        ax2.view_init(elev=0, azim=-90)

        ax3.scatter(x, y, z, c=colors, alpha=0.6, s=20)
        ax3.set_title('Side View (0°, 0°)', fontsize=10)
        ax3.view_init(elev=0, azim=0)

        ax4.scatter(x, y, z, c=colors, alpha=0.6, s=20)
        ax4.set_title('Top View (90°, -90°)', fontsize=10)
        ax4.view_init(elev=90, azim=-90)

        ax5.scatter(x, y, z, c=colors, alpha=0.6, s=20)
        ax5.set_title('Alternate View (45°, 45°)', fontsize=10)
        ax5.view_init(elev=45, azim=45)

        ax6.scatter(x, y, z, c=colors, alpha=0.6, s=20)
        ax6.set_title('Downhill (-30°, 120°)', fontsize=10)
        ax6.view_init(elev=-30, azim=120)

    for ax in [ax1, ax2, ax3, ax4, ax5, ax6]:
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')

        ax.grid(True, alpha=0.3)
```

```python
fig = plt.figure(figsize=(16, 12))
plot_clusters(fig, [np.array(c.data) for c in sorted_data])

plt.tight_layout()
plt.suptitle('3D Visualization - Multiple Perspectives', fontsize=14, y=1.02)
plt.show()
```
